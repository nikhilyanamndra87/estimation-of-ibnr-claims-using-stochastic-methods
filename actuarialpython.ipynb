```python
pipe = cl.Pipeline(steps=[
    ('dev',cl.Development()),
    ('chainladder',cl.Chainladder())])

# Develop scoring function that returns an Ultimate/Incurred Ratio

# Set up a GridSearch space
grid = cl.GridSearch(
    estimator=pipe,
    param_grid=dict(
        dev__n_periods=[item for item in range(2,11)],
        dev__average=['simple', 'volume', 'regression']),
    scoring=lambda x: (
        x.named_steps.chainladder.ultimate_.sum() /
        motor_triangle.latest_diagonal.sum()))
grid.fit(motor_triangle)

# Plot data
results = pd.pivot_table(grid.results_, index='dev__n_periods',
                         columns='dev__average', values='score')
```

```python
import chainladder as cl

# Simulation parameters
random_state = 42
n_sims = 1000

# Get data
loss = motor_triangle
premium = loss.latest_diagonal * 0 + 8e6

# Simulate loss triangles
sim = cl.BootstrapODPSample(random_state=random_state, n_sims=n_sims)
sim.fit(loss, sample_weight=premium)


# Fit Bornhuetter-Ferguson to stochastically generated data
model = cl.BornhuetterFerguson(0.65, apriori_sigma=0.10)
model.fit(sim.resampled_triangles_, sample_weight=premium)

# Grab completed triangle replacing simulated known data with actual known data
full_triangle = (model.full_triangle_ - model.X_ + loss) / premium

# Limiting to the current year for plotting
current_year = full_triangle[full_triangle.origin==full_triangle.origin.max()].to_frame().T
```

```python

def calculate_residuals(model, features, label):
    predictions = model.predict(features)
    df_results = pd.DataFrame({'Actual': label.squeeze(1), 'Predicted': predictions})
    df_results['Residuals'] = abs(df_results['Actual']) - abs(df_results['Predicted'])
    df_results['Residuals_per'] = (abs(df_results['Actual']) - abs(df_results['Predicted']))/df_results['Actual']
    
    return df_results
```
